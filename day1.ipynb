```julia
function extract_numbers_to_array(input_file::AbstractString)
    numbers_array = Vector{String}()

    open(input_file, "r") do input
        for line in eachline(input)
            numbers = filter(x -> isdigit(x) || x == '.', line)
            
            push!(numbers_array, join(numbers))
        end
    end

    return numbers_array
end

numbers_array = extract_numbers_to_array(input_file)

for (i, numbers) in enumerate(numbers_array)
    println("Line $i: $numbers")
end
```

```julia
# Read input from file
filename = "puzzle1.txt"
file = open(filename, "r")
s = read(file, String)
close(file)

# Initialize counter
r = 0

# Process each line in the input
for x in split(strip(s), '\n')
    fst = nothing
    lst = nothing

    # Process each character in the line
    for y in x
        if isdigit(y)
            lst = y
            fst === nothing && (fst = y)
        end
    end

    # Update counter
    r += parse(Int, string(fst) * string(lst))
end

println(r)

# Define a dictionary for mapping number words to digits
m = Dict(
    "zero" => 0,
    "one" => 1,
    "two" => 2,
    "three" => 3,
    "four" => 4,
    "five" => 5,
    "six" => 6,
    "seven" => 7,
    "eight" => 8,
    "nine" => 9
)

r = 0

# Process each line in the input (resetting variables)
for x in split(strip(s), '\n')
    fst = nothing
    lst = nothing
    s = ""

    # Process each character in the line
    for c in x
        dig = nothing

        if isdigit(c)
            dig = c
        else
            s *= c

            for (k, v) in m
                if endswith(s, k)
                    # process as digit v
                    dig = string(v)
                end
            end
        end

        if dig !== nothing
            lst = dig
            fst === nothing && (fst = dig)
        end
    end

    # Update counter
    r += parse(Int, string(fst) * string(lst))
end

println(r)
```
